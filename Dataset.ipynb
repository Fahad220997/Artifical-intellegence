```python
from os import path
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

path = "/content/youtube-top-100-songs-2025.csv"

df = pd.read_csv(path)
print(df.description)


```

```python
channel_followers = df.groupby('channel')['channel_follower_count'].max().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 5))
channel_followers.plot(kind='bar', color='teal')
plt.title("Top 10 YouTube Channels by Follower Count")
plt.xlabel("Channel")
plt.ylabel("Follower Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
